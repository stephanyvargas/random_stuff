# Create a machine learning model to predict prices from Boston house price data

The standard data set for data analysis practice, "Boston housing price data," has been compiled into Excel.

For 490 residences in Boston, crime rate by town (CRIM), distance to employment center (DIS), student-teacher ratio (PTRATIO), etc.

Conditions that are likely to affect housing prices and actual prices are summarized.

There are two sheets in the Excel file.

The “For ML” sheet shows the conditions and prices of 490 houses.

The 'For Prediction' sheet contains the conditions of the house you want to predict (price is unknown).

Let's create a machine learning model that predicts house prices based on For ML data and predict prices based on For Prediction conditions.

You can get the Excel file [here](https://github.com/hatanaka-lab/Getting_started_with_MI/tree/main/data/Boston.xlsx).

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## STEP1: Load data

```python
# Read data from "For ML" Sheet of Excel file and assign to df
# df stands for "data frame".
# In Python, two types of data frames are often used: "pandas" data frames and "numpy" arrays, so it is a good idea to get into the habit of naming data frames df to avoid confusion.
df = pd.read_excel("Boston.xlsx", sheet_name='For ML')
```

```python
# Check the contents of df①
# Check the name and number of data in each column (no gaps, etc.)
df.info()
```

```python
df.head()
```

## STEP2: Observe the data

```python
# Observing data distribution: Part 1
df_Target=df.drop(columns=['No'])
pd.plotting.scatter_matrix(df_Target, 
                           figsize=(12, 12), 
                           hist_kwds={'bins':10},
                           marker=('o'), s=8, 
                           alpha=.5)
plt.show()
```

```python
# Observe the data distribution: Part 2 (correlation matrix)
correlation_coefficients = df_Target.corr() # Calculate correlation matrix
# Heatmap of correlation matrix (with correlation coefficient values)
plt.rcParams['font.size'] = 9
plt.figure(figsize=(12, 10)) # Specify the image size at this stage
sns.heatmap(correlation_coefficients, vmax=1, vmin=-1, cmap='seismic', square=True, annot=True, xticklabels=1, yticklabels=1)
plt.xlim([0, correlation_coefficients.shape[0]])
plt.show()
```

## STEP3: Separate the data

```python
# Assign columns 1 to 14 of df to X (used as explanatory variable)
# The number of columns is counted from 0. So No is column 0, CRIM is column 1, LSTAT is line 13
X = np.array(df.iloc[:,1:14])
df.iloc[:,1:14].head(2) #For checking the contents
```

```python
# Assign the column labeled Price to Y (use it as the objective variable)
Y = np.array(df["Price"])
df["Price"].head(2) #For checking the contents
```

```python
# Standardize explanatory variables
# Standardized value = (original value - mean value)/standard deviation
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X)
X_scaled = scaler.transform(X)
```

```python
# Divide the data into training data (train) and verification data (test) for holdout verification.
# The division is random (this time, it is divided into train : test = 80% : 20%)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, Y, test_size=0.2,random_state=999)
```

## STEP4: Determine hyperparameters

```python
# Decide hyperparameters for RandomForest
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score

list_param = []
list_score = []

for ne in range(5,250,50):
    for nd in range(5,41,5):
        model = RandomForestRegressor(n_estimators=ne, max_depth=nd, random_state=555)
        cv5_score = cross_val_score(model, 
                                    X_train, 
                                    y_train, 
                                    cv=5).mean() #5-fold CV using training data (train)
        print("num_trees=",ne,"max_depth=",nd,"R2_score=",cv5_score)
        list_param.append([ne,nd])
        list_score.append(cv5_score)
        
max_index = np.argmax(list_score)
print("")
print("----Best parameters-----")
print("num_trees=",list_param[max_index][0],
      "max_depth=",list_param[max_index][1],
      "R2_score=",list_score[max_index])
```

## STEP5: Machine learning execution

You can try various methods by replacing the RandomForestRegressor part that appears immediately after this with the name of another machine learning method.

```python
# Machine learning execution
# Adopt Random Forest regression
# When you run this, the contents of "model" will become a machine learning model created from the training data
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators=155,
                              max_depth=25,
                              random_state=1)
model.fit(X_train, y_train)
```

```python
# Verify the validity of the machine learning model
# Check the scores of training data and test data (the closer the score is to 1, the better the model)
print("Score for Training Data:", model.score(X_train,y_train))
print("Score for Test Data :", model.score(X_test ,y_test ))
```

```python
# draw a scatter plot
# The horizontal axis is the actual value (y): the vertical axis is the value predicted from the machine learning model
# The more the data is on the diagonal, the higher the prediction accuracy
plt.figure(figsize=(6,6))
plt.scatter(y_train,model.predict(X_train),c='b',marker='o',alpha=0.7,label='Train')
plt.scatter(y_test, model.predict(X_test ),c='r',marker='^',alpha=0.7,label='Test')
# draw a diagonal line
x1 = np.linspace(-1, 50, 100)
plt.plot(x1, x1, linestyle='-',c="silver")
#
plt.title("Random Forest Regression", fontsize=14)
plt.ylabel("Predicted value", fontsize=12)
plt.xlabel("Actual value", fontsize=12)
plt.legend(fontsize=12)
plt.show()
```

Horizontal axis: actual value, vertical axis: predicted value by machine learning. The more the data is on the diagonal line, the better the prediction accuracy is.

## STEP 6: Prediction for data with unknown house prices

```python
# Calculate the Price value by inputting it to the machine learning model
model.predict(X_Pred_scaled)
```

## Bonus: Consider various other machine learning models

### Example 1: PLS regression

```python
# Decide hyperparameters for PSL
from sklearn.cross_decomposition import PLSRegression
from sklearn.model_selection import cross_val_score

list_param = []
list_score = []

for k in range(2,13,1):
    model = PLSRegression(n_components=k)
    cv5_score = cross_val_score(model, 
                                X_train, 
                                y_train, 
                                cv=5).mean() #5-fold CV using training data (train)
    print("n_components=", k, "R2_score=", cv5_score)
    list_param.append(k)
    list_score.append(cv5_score)
    
max_index = np.argmax(list_score)
print("")
print("----Best parameters-----")
print("n_components=",list_param[max_index],"R2_score=",list_score[max_index])
```

```python
# Select machine learning method
# This time PLS regression (personal favorite)
from sklearn.cross_decomposition import PLSRegression

model = PLSRegression(n_components=8)

# Machine learning execution: 
model.fit(X_train, y_train)

# verification
print("Score for Training Data:", model.score(X_train,y_train))
print("Score for Test Data :", model.score(X_test ,y_test ))

# Scatter plot drawing
plt.figure(figsize=(6,6))
plt.scatter(y_train,model.predict(X_train),
            c='b',marker='o',
            alpha=0.7,label='Train')
plt.scatter(y_test, model.predict(X_test),
            c='r',marker='^',
            alpha=0.7,label='Test')
x1 = np.linspace(-1, 50, 100)
plt.plot(x1, x1, linestyle='-',c="silver")
plt.title("PLS regression", fontsize=14)
plt.ylabel("Predicted value", fontsize=12)
plt.xlabel("Actual value", fontsize=12)
plt.legend(fontsize=12)
plt.show()
```

### Example 2: LASSO regression

```python
# Select machine learning method
# Perform both hyperparameter tuning and regression for LASSO regression (kindly♪)
from sklearn.linear_model import LassoCV

model = LassoCV(n_alphas=50, cv=5, max_iter=100000)

# Machine learning execution: 
model.fit(X_train, y_train)

# Display the hyperparameter (α) determined by 5-fold CV
print('alpha = ', model.alpha_)

# verification
print("Score for Training Data:", model.score(X_train,y_train))
print("Score for Test Data :", model.score(X_test ,y_test ))

# Scatter plot drawing
plt.figure(figsize=(6,6))
plt.scatter(y_train,model.predict(X_train),c='b',marker='o',alpha=0.7,label='Train')
plt.scatter(y_test, model.predict(X_test ),c='r',marker='^',alpha=0.7,label='Test')
x1 = np.linspace(-1, 50, 100)
plt.plot(x1, x1, linestyle='-',c="silver")
plt.title("LASSO regression", fontsize=14)
plt.ylabel("Predicted value", fontsize=12)
plt.xlabel("Actual value", fontsize=12)
plt.legend(fontsize=12)
plt.show()
```
